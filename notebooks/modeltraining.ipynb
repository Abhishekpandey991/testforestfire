```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df.head()
```

```python
##drop month,day and year
df.drop(columns=['day','month','year'],inplace=True)
```

```python
df.columns
```

```python
df.head()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df['Classes'].value_counts()
```

```python
df.head()
```

```python
#Independent Value and dependent variable
X=df.drop(columns='FWI')
y=df['FWI']
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
correlation_features=correlation(X_train,0.85)
```

```python
X_train.drop(correlation_features,axis=1,inplace=True)
X_test.drop(correlation_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scalaed=scaler.transform(X_test)
```

```python
X_train_scaled,X_test_scalaed
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
linereg=LinearRegression()
linereg.fit(X_train_scaled,y_train)
y_pred=linereg.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

### Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
y_pred=lassocv.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

## Ridge Regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgeCV=RidgeCV(cv=5)
ridgeCV.fit(X_train_scaled,y_train)
y_pred=ridgeCV.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

```python
ridgeCV.get_params()
```

## Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
elasticNet=ElasticNet()
elasticNet.fit(X_train_scaled,y_train)
y_pred=elasticNet.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNetCV
elasticnetcv=ElasticNetCV(cv=5)
elasticnetcv.fit(X_train_scaled,y_train)
y_pred=elasticnetcv.predict(X_test_scalaed)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
plt.scatter(y_test,y_pred)
```
